```python
age_count = data['Age'].value_counts().sort_index()

plt.bar(age_count.index, age_count.values, color='pink', edgecolor='black')
plt.title('Distribution of generations', fontsize=10)
plt.xlabel('Ages')
plt.ylabel('Count')
plt.gcf().patch.set_facecolor('green')
plt.gca().set_facecolor('yellow')
plt.show()
```

```python
grouped_columns = ['Age', 'Monthly Income', 'Family size', 'Feedback']
```

```python
grouped_data = data.groupby(grouped_columns).size().reset_index(name='count')
grouped_data
```

```python
import pandas as pd
```

```python
gro_dfz = data.groupby(['Age','Feedback']).size().reset_index(name='count')
gro_dfz
```

```python
heatmap_data = gro_dfz.pivot('Age', 'Feedback', 'count')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True)
plt.title('Feedback Counts by Age')
plt.xlabel('Feedback')
plt.ylabel('Age')
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_counts = data['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```
