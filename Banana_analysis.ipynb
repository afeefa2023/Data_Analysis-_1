```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Bananas receive the same count of good and bad.
sns.countplot(x='Quality', data=df)
plt.title('Distribution of Banana Quality ')
plt.show()
```

```python
#Most of the relationship between weight and size in bananas
sns.scatterplot(data=df, x="Size", y="Weight", color='pink')
plt.show()
```

```python
sns.lineplot(data=df, x='Size', y="Sweetness", hue="Softness", style="Quality", markers=True)
plt.show()
```

```python
df.corr()
```

```python
sns.barplot(df)
```

```python
sns.pairplot(df)
plt.title('Pairplot of Banana Data')
plt.show()
```

```python
#Bananas are getting longer and more frequent at the same time.
sns.histplot(df['Size'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Banana Length')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Size', y='Softness', data=df)
plt.title('Relationship between size and softness')
plt.xlabel('size')
plt.ylabel('weight')
plt.show()
```
