```python
mean_price = data['Price'].mean()
std_price = data['Price'].std()
car_data = data[~((data['Price'] - mean_price) > 3 * std_price)]
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(car_data['Price'], bins=22,kde=True)
plt.title('The Dispersion of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
