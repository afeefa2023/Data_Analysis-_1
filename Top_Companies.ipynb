```python
dz.fillna(1, inplace=True)
dz
```

```python
missing_values = dz.iloc[:, :10].isnull().sum()

missing_values
```

```python
companies_10=dz.head(10)
```

```python
#Accenture (Co.), the top-ranked business in this
plt.figure(figsize=(10, 6))
sns.barplot(x='Ratings', y='Company_name', data=companies_10, palette='viridis')
plt.title('Company Ratings for Top 10 Companies')
plt.xlabel('Ratings')
plt.ylabel('Company Name')
plt.show()
```

```python
#The majority of companies with ratings below 0.5
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(companies_10['Ratings'], bins=20, kde=True, color='pink')
plt.title('Distribution of Company Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
import random
```

```python
rand = dz.sample(n=10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Ratings', y='Company_name', data=rand, palette='viridis')
plt.title('Ratings of Random Companies')
plt.xlabel('Ratings')
plt.ylabel('Company Name')
plt.show()
```

```python
sort= dz.sort_values(by='Ratings', ascending=False)
```

```python
#Under the 10-point scale, all companies obtain the same rating.
top_companies = sort.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Ratings', y='Company_name', data=top_companies, palette='viridis')
plt.title('Ratings of Top Companies')
plt.xlabel('Ratings')
plt.ylabel('Company Name')
plt.show()
```

```python
sort = dz.sort_values(by='Ratings', ascending=False)
```

```python
top_companies = sort.head(10)
```

```python
#between Dr. SNS and Teqtous, with a majority rating of 10 and the total number of jobs availab
plt.figure(figsize=(12, 8))
sns.barplot(x='Ratings', y='Company_name', data=top_companies, palette='viridis', label='Ratings')
sns.barplot(x='Total_jobs_available', y='Company_name', data=top_companies, color='orange', label='Total Jobs Available')

plt.title('Ratings and Total Jobs Available for Top Companies')
plt.xlabel('Count')
plt.ylabel('Company Name')
plt.legend()
plt.show()
```
