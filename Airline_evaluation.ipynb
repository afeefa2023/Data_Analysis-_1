```python
plt.figure(figsize=(12, 6))

gr_data= data.groupby('Airline')['Overall Rating'].mean().reset_index()

colors = ['skyblue', 'salmon', 'lightgreen']

result = plt.bar(gr_data['Airline'], gr_data['Overall Rating'], color=colors)

for bar in result:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             round(bar.get_height(), 2), ha='center', va='bottom')

plt.xlabel("Airline")
plt.ylabel("The mean total rating")
plt.title("overall ratings for Airlines")
plt.show()
```

```python
grouped_data = data.groupby('Class')['Overall Rating'].mean().reset_index()

```

```python
grouped_data.reset_index(drop=True, inplace=True)

heatmap_data = grouped_data.pivot( columns='Class', values='Overall Rating')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt=".2f")
plt.xlabel("Class")
plt.ylabel("rating")
plt.title("overall ratings for Heatmap")
plt.show()
```
