```python
gr_data = data.groupby('CSAT Score')['Unique id']
```

```python
count_per_score = gr_data.count()
count_per_score
```

```python
gr_data = data.groupby('CSAT Score')['Unique id'].count().reset_index()

```

```python
heatmap_data = gr_data.pivot(index='CSAT Score', columns='Unique id', values='Unique id')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap="pink", annot=True, fmt="g")
plt.xlabel("Unique id")
plt.ylabel("CSAT Score")
plt.title("Heatmap of CSAT Score vs Unique ID")
plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.histogram(data, x='CSAT Score', nbins=30,
                   title='Score Distribution for CSAT',
                   color_discrete_sequence=['pink'])
fig.update_layout(xaxis_title='CSAT Score', yaxis_title='Frequency')
fig.show()
```

```python
channel_counts = data['channel_name'].value_counts()
```

```python
custom_colors = ['#FF69B4', '#FFFF00', '#FFA500']
```

```python
fig = px.pie(values=channel_counts.values, names=channel_counts.index, title='dispersion of the channel',
 color_discrete_sequence=custom_colors)

fig.show()
```
