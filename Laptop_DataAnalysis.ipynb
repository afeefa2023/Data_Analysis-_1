```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('/content/archive (6).zip')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop_duplicates()
```

```python
value_counts = data['brand'].value_counts()

value_counts
```

```python
value_counts = data['Rating'].value_counts()

value_counts
```

```python
counts = data.groupby(['brand', 'Price', 'primary_storage_capacity']).size().reset_index(name='count')

counts
```

```python
data.select_dtypes(include=['object']).nunique()
```

```python
data.shape
```

```python
brand_price = data.groupby('brand')['Price'].mean().sort_values(ascending=False)
brand_price
```

```python
sorted_data = data.sort_values(by='Price')
```

```python
sns.histplot(data['Price'], kde=True, bins=20, color='skyblue')
plt.title('Changes in pricing')
plt.xlabel('variance')
plt.ylabel('frequency')
plt.show()
```

```python
sns.boxplot(x='Price', y='Rating', data=data,)
plt.title('Relationship between Price and Rating')
plt.xlabel('price')
plt.ylabel('rating')
plt.show()
```

```python
OS_Coun = data['OS'].value_counts()
OS_Coun
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(data = data, x = 'OS',color='pink')
plt.title(' Operating system Available')
plt.xlabel('OS Types')
plt.ylabel = ('Count')
plt.show()
```
